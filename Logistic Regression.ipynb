```python
from sklearn import datasets
import numpy as np
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
iris.data
```

```python
x = iris["data"][:,3:]
```

```python
x
```

```python
y = iris["target"] == 2
y
```

```python
y = (iris["target"] == 2).astype(np.int) ## true is 1 false is 0
y
```

```python
## TRAIN A LOGISTIC CLASSIFIER
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(x,y)
```

```python
clf.predict([[1.6]])
```
