```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
dframe1 = DataFrame({'key':['x','y','z','z','x','x'], 'data_set_1':np.arange(6)})
dframe1
```

```python
dframe2 = DataFrame({'key':['q','y','z'], 'data_set_2':[1,2,3]})
dframe2
```

```python
pd.merge(dframe1, dframe2)
```

```python
dframe3 = DataFrame({'data_set_1':np.arange(6), 'key':['x','z','y','z','x','x']})
dframe3
```

```python
dframe4 = DataFrame({'data_set_2':[1,2,3], 'key':['q','y','z']})
dframe4
```

```python
pd.merge(dframe3, dframe4)
```

```python
pd.merge(dframe3, dframe4, on = 'key', how = 'left')
```

```python
pd.merge(dframe3, dframe4, on = 'key', how = 'right')
```

```python
dframe3
```

```python
dframe4
```

```python
pd.merge(dframe3, dframe4, on = 'key', how = 'outer')
```

```python
dleft = DataFrame({
    'key1': ['SF', 'SF', 'LA', 'LA'],
    'key2': ['one', 'one', 'one', 'two'],
    'left_data':[40,50,60,70]
})
```

```python
dright = DataFrame({
    'key1':['SF', 'SF', 'LA', 'LA'],
    'key2':['one', 'one', 'one', 'two'],
    'left_data' : [40,50,60,70]
})
```

```python
dleft
```

```python
dright
```

```python
pd.merge(dleft, dright, on=['key1', 'key2'], how = 'outer')
```

```python
pd.merge(dleft, dright, on = 'key1', suffixes = ('_lefty', '_right'))
```
